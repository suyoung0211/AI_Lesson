```python
# ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ ì •ì˜
class Counter:
    def __init__(self):
        self.num = 0     # ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ num ê°’ì„ 0ìœ¼ë¡œ ì´ˆê¸°í™”
    # ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ : ì•„ë˜ 4ê°œ
    # ì¸ìŠ¤í„´ìŠ¤ ê°ì²´ë¥¼ ë§Œë“¤ì–´ì„œ ì‹¤í–‰í•˜ëŠ” ë©”ì†Œë“œ
    def increment(self):
        self.num += 1
    def current_value(self):
        return self.num
    def decrement(self):
        self.num -= 1
    def reset(self):
        self.num = 0
```

```python
# ê°ì²´(ì¸ìŠ¤í„´ìŠ¤) ìƒì„±
counter = Counter()

# increment ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ ì‹¤í–‰
counter.increment()
counter.increment()
counter.increment()
print(counter.num)

# reset ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ ì‹¤í–‰
counter.reset()
print(counter.num)

# decrement ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ ì‹¤í–‰
counter.decrement()
counter.decrement()
print(counter.num)
```

- í´ë˜ìŠ¤ ì •ì˜ì‹œ ì‚¬ìš©í•˜ëŠ” ë°ì½”ë ˆì´í„°(ìë°”ì˜ ì• ë…¸í…Œì´ì…˜)


```python
class Cat:
    def __init__(self):
        self.__age = 0
        # __ë³€ìˆ˜ëª… : ë‚´ë¶€ ë³€ìˆ˜ ì§ì ‘ ì ‘ê·¼ ë°©ì§€. ìº¡ìŠí™”

    @property    # ë©”ì„œë“œë¥¼ ì†ì„±ì²˜ëŸ¼ ì ‘ê·¼í•  ìˆ˜ ìˆê²Œ í•´ì¤Œ. getter í˜•ì‹
    def age(self):
        return self.__age

    @age.setter   # setter : @ì†ì„±.setter
    def age(self, value):
        if value >= 0:
            self.__age = value
        else:
            raise ValueError("ë‚˜ì´ëŠ” ìŒìˆ˜ê°€ ë  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

```python
lucky = Cat()
lucky.age = 10   # setter ê°€ ìˆì„ ë•Œë§Œ ê°€ëŠ¥ . setter ì—†ìœ¼ë©´ ì˜¤ë¥˜
print(lucky.age)
```

```python
# ì¸ìŠ¤í„´ìŠ¤ a,b ë¥¼ ìº¡ìŠí™”í•˜ê³  getter,setter ìœ„ì˜ ë°ì½”ë ˆì´í„°ë¡œ ì‘ì„±í•´ë³´ê¸°
class MyMath:
    def __init__(self, a=0, b=0):
        self.__a = a
        self.__b = b

    @property    # property ì•„ë˜ í•¨ìˆ˜ ì´ë¦„ ì†ì„± ì´ë¦„ì´ ë©ë‹ˆë‹¤.
    def first(self):
      return self.__a

    @property
    def second(self):
      return self.__b

    @first.setter
    def a(self,value):
      self.__a=value

    @second.setter
    def b(self,value):
      self.__b=value

# ê°€ëŠ¥í•œ ì¼€ì´ìŠ¤ì´ì§€ë§Œ getter ì™€ setter í•¨ìˆ˜ ì´ë¦„ì€ í†µì¼ ì¶”ì²œ.
math1 = MyMath()
math1.a = 23   # setter
math1.b = 32   # setter
print(f'{math1.first}, {math1.second}')
```

```python
class MyMath:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def substract(self):
        return self.a - self.b

    def multiply(self):
        return self.a * self.b

    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Error: Division by zero"
```

```python
class MyMath:
    def __init__(self, first=0, second=0):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def substract(self):
        return self.first - self.second

    def multiply(self):
        return self.first * self.second

    def divide(self):
        if self.second != 0:
            return round(self.first / self.second, 2)
        else:
            return "Error: Division by zero"
    def __str__(self):
        return f'first : {self.first}, second : {self.second}'
```

```python
# MyMath í´ë˜ìŠ¤
# ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ : first , second
# ìƒì„±ì : first,second ë¥¼ ì´ˆê¸°í™”. ê¸°ë³¸ê°’ì€ 0
# __str__ ë©”ì†Œë“œ : first, second ì˜ ê°’ì„ ì¶œë ¥
# ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ : add,substract, multiply, divide ê³„ì‚°ê²°ê³¼ê°’ ë¦¬í„´

mymath = MyMath();
print(mymath)
mymath = MyMath(34,12);
print(mymath)
print(mymath.add())
print(mymath.substract())
print(mymath.multiply())
print(mymath.divide())
```
