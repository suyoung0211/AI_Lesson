## enumerate() í•¨ìˆ˜

 * Iterable(ë°˜ë³µ ê°€ëŠ¥ ê°ì²´): ë¦¬ìŠ¤íŠ¸, íŠœí”Œ, ë”•ì…”ë„ˆë¦¬, ë¬¸ìì—´ê³¼ ê°™ì€ ì»¬ë ‰ì…˜ ìë£Œí˜•ì€ ë°˜ë³µ ê°€ëŠ¥í•œ ê°ì²´ë¡œ, ë°˜ë³µìë¥¼ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
 * Iterator(ì´í„°ë ˆì´í„°): ë°˜ë³µì(è€…). ì´í„°ëŸ¬ë¸” ê°ì²´ì—ì„œ iter() í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•´ ì–»ì„ ìˆ˜ ìˆìœ¼ë©°, next() í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ìš”ì†Œë¥¼ í•˜ë‚˜ì”© ìˆœì°¨ì ìœ¼ë¡œ ì ‘ê·¼í•©ë‹ˆë‹¤.

### zip() í•¨ìˆ˜
- zip() í•¨ìˆ˜ëŠ” íŒŒì´ì¬ì—ì„œ ì—¬ëŸ¬ ê°œì˜ ë°˜ë³µ ê°€ëŠ¥í•œ(iterable) ê°ì²´ë¥¼ ë³‘ë ¬ì ìœ¼ë¡œ ë¬¶ì–´ì£¼ëŠ” ë‚´ì¥ í•¨ìˆ˜ì…ë‹ˆë‹¤. ê° ê°ì²´ë¡œë¶€í„° ìˆœì„œëŒ€ë¡œ í•˜ë‚˜ì˜ ìš”ì†Œì”© ê°€ì ¸ì™€ `íŠœí”Œë¡œ ë¬¶ì–´ì£¼ëŠ”` ì´í„°ë ˆì´í„°ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤. ì£¼ë¡œ ì—¬ëŸ¬ ê°œì˜ ë¦¬ìŠ¤íŠ¸ë‚˜ íŠœí”Œì„ ë³‘ë ¬ì ìœ¼ë¡œ ì²˜ë¦¬í•˜ê³ ì í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.
- zip(iterable1, iterable2, ...) : 2ê°œ ì´ìƒì˜ ë°˜ë³µ ê°ì²´ë¥¼ ëŒ€ìƒìœ¼ë¡œ í•©ë‹ˆë‹¤.

```python
test1 = [100,101,102,103]
test2 = ['apple','banana','orange','melon']

# zip í•¨ìˆ˜ ì—†ì´ forë¬¸ í•˜ë‚˜ë¡œ 2ê°œì˜ ë¦¬ìŠ¤íŠ¸ì—ì„œ ìˆœì°¨ì  ê°’ ê°€ì ¸ì˜¤ê¸°
# ìë°” : for(i=0;i<test1.size();i++) . íŒŒì´ì¬ì€ X
for i in range(len(test1)):
  print(test1[i],test2[i])
# test1 ì˜ ê°¯ìˆ˜ê°€ ë” ë§ìœ¼ë©´ ì˜¤ë¥˜
```

```python
test1 = [100,101,102,103]
test2 = ['apple','banana','orange','melon']
for item in zip(test1,test2):
  print(item)
```

```python
# ìœ„ì—ì„œ item ì€ tuple
# ê°ê° ë°ì´í„°ë¥¼ ì ‘ê·¼í•˜ê¸° ìœ„í•´ì„œëŠ” ì–¸íŒ©í‚¹ì„ ë” ë§ì´ ì‚¬ìš©
for t1,t2 in zip(test1,test2):
  print(t1, t2)
```
