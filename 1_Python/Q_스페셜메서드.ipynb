### __ 로 시작하고 __ 로 끝나는 메서드
- 파이썬의 스페셜 메서드 (또는 매직 메서드라고도 불림) 는 더블 언더스코어(__) 로 시작하고 끝나는 메서드 이름을 갖습니다.
- 특징
  1. 객체 내부에서 정의 : 스페셜 메서드는 클래스 내부에서 정의됩니다.
  2. 특정 연산을 오버라이드 : object 클래스에서 기본적으로 제공되는 메서드를 재정의하여 커스터마이징할 수 있도록 합니다. 예를 들어 repr, str, add, len 등이 있습니다.
  3. 간접호출 : 스페셜 메소드는 특정 구문이나 내장함수를 사용할 때 파이썬 인터프리터에 의해 자동으로 호출됩니다.

### 1. __ repr __()
- "representation". 파이썬에서 객체의 "공식적인" 문자열 표현을 정의하는 특별한 메서드로, 주로 개발자나 디버깅을 위한 정보를 제공하는데 사용됩니다. 이 메서드는 객체를 문자열로 변환할 때 자동으로 호출됩니다.
- 또는 repr() 함수를 통해 호출

````python
## __repr__() 스페셜 메소드와 eval() 함수

class Dog:
    def __init__(self, name, age=0):
        self.name = name
        self.age = age

    # def __repr__(self):
      #  return f"{self.__class__.__name__}(name='{self.name}', age={self.age})"


coco = Dog('코코', 15)
print(coco)      # __str__ 또는 __repr__ 정의한 리턴 내용
print(repr(coco))           # 함수로 호출
print(coco.__repr__())      # 객체 직접 호출
````

- `__repr__` 재정의 하면

````python
class Dog:
    def __init__(self, name, age=0):
        self.name = name
        self.age = age

    def __repr__(self):
       return f"{self.__class__.__name__}(name='{self.name}', age={self.age})"


coco = Dog('코코', 15)
print(coco)      # __str__ 또는 __repr__ 정의한 리턴 내용
print(repr(coco))           # 함수로 호출
print(coco.__repr__())      # 객체 직접 호출
````

- eval() 함수

````python
# eval()은 주어진 '문자열을 파이썬 표현식으로 평가하고 실행'하여 그 결과를 반환하는 내장 함수

x = 10
y = 3
result = x + y
print(result)
result = eval("x + y")
print(result)
eval("print('hello😁')")
````

````python
coco_repr = "Dog[name='바둑이', age=23]"   # __repr__ 메소드의 기본값
# repr(coco) 리턴 즉 __repr__ 메소드의 리턴 문자열 형식과 동일함
coco_repr= repr(coco)
eval_coco = eval(coco_repr)

print(type(eval_coco))
print(eval_coco.name)
print(eval_coco.age)
print(type(coco_repr))
print(type(eval_coco))
````

### 2. __ str __()
- 사용자가 이해하기 쉽고 읽기 좋은 형태의 문자열 표현을 반환하는 메서드로, 주로 객체를 출력할 때 사용됩니다.
- __ str __ 는 print() 나 str() 를 통해 호출됩니다.
- repr() 는 객체의 공식적인 표현을 제공하고, repr() 함수를 통해 호출됩니다.

````python
# __str__ 과 __repr__ 비교
class Book:
  def __init__(self,title,writer):
    self.title=title
    self.writer=writer

  # str 함수를 오버라이딩하기
  def __str__(self):
   return f"{self.title} : {self.writer}"

  def __repr__(self):
    return f"Book(title='{self.title}', writer='{self.writer}')"

book = Book('미친듯이 재밌는 자바','김땡땡')
print(book)   # 오버라이딩 한 리턴값을 출력하기. repr과 str 중 str 실행
print(str(book))
print(repr(book))
````

````python
# 리스트는 __repr__ , __str__ 메소드를 재정의 하여 요소의 목록을 문자열로 리턴함.
li=[1,2,3,4,5]
print('li =',li)
print(type(li))
print('repr(li) =',repr(li))
print(type(repr(li)))
print('str(li) =',str(li))
print(type(str(li)))
````

### 4. __ getitem __()
- 파이썬에서 인덱싱을 지원하기 위해 사용되는 특별한 메서드
- 객체의 특정 항목을 가져오기 위해 대괄호([])를 사용할 때 호출됩니다.
- 인덱스나 키에 해당하는 값을 반환하는 역할을 합니다.

````python
# __getitem__() : 객체의 인스턴스변수(속성)을 인덱스로 접근할 수 있게 한다.
class Point:
    def __init__(self, x, y,z):
        self.x = x
        self.y = y
        self.z = z

    def __getitem__(self, index):
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        elif index == 2:
            return self.z
        else:
            return -1
pt = Point(5, 3, 7)
print(pt)
print(pt[0])
print(pt[1])
print(pt[2])
print(pt[100])
````

````python
class MyList:
    # 파이썬은 동적 타입 언어. data 속성의 타입은 객체가 만들때 결정.
    def __init__(self, data):
        self.data = data
    # def __getitem__(self, index):
    #     return self.data[index]

# data 속성의 타입을 list 로 할 것이다.
ml = MyList([10, 20, 30, 40])
# ml 은 MyList 객체를 참조합니다. 이 변수는 속성 data 접근이 가능합니다.
print(ml.data)
print(ml.data[0])
print(ml.data[1])
# print(ml.data[5])   # 오류
````

````python
class MyList:
    def __init__(self, data):
        self.data = data
    def __getitem__(self, index):
        return self.data[index]

# __getitem__ 메소드로 data 의 특정 인덱스값을 리턴
#  ↪ data 일때만 사용 가능.
ml = MyList([10, 20, 30, 40])
print(ml[0])
print(ml[1])
print(ml[-1])
````

### 5. __ call __()
- 파이썬에서 객체를 함수처럼 호출할 수 있게 해주는 특별한 메서드입니다.
- 해당 클래스의 인스턴스를 함수처럼 사용할 수 있습니다.
- 인스턴스에 대해 괄호() 를 사용하여 값을 전달하고 결과를 반환 받을 수 있습니다.

````python
# __call__() 메소드 : 객체를 참조하는 변수를 함수처럼 사용할 수 있게 한다.
#  ↪ 아래 예시는 __call__ 메소드가 action 인자를 사용합니다.
````

````python
class Dog:
  def __init__(self, name, family):
    self.name = name
    self.family = family
  def __call__(self, action):
    return f"{self.name}, the {self.family}, is now {action}!"

# Dog 객체 생성
my_dog = Dog("Buddy", "Golden Retriever")
print(my_dog)

# __call__ 메소드 호출 . 인자값 action 전달함.
print(my_dog("running"))  # Buddy, the Golden Retriever, is now running!
print(my_dog("barking"))  # Buddy, the Golden Retriever, is now barking!
````

### 6. __ add __() 메서드 (연산자 오버로딩)
- 이 코드에서 add()는 연산자 오버로딩(Operator Overloading) 을 통해 + 연산자를 사용자 정의하는 역할을 합니다.

````python
# __add__() 메소드 : 객체들의 덧셈 연산을 재정의
# __mul__() 메소드 : 객체들의 곱셈 연산을 재정의
class MyCalculator:
    def __init__(self, num):
        self.num = num  # 숫자 값을 저장

    def __add__(self, other):
        return self.num + other.num  # 두 객체의 num 값을 더함

    def __mul__(self, other):
        return self.num * other.num  # 두 객체의 num 값을 곱함


# 객체 생성
c1 = MyCalculator(10)    # 이 객체의 num 변수값이 10
c2 = MyCalculator(20)
c3 = MyCalculator(30)

# + 연산 수행
result = c1 + c2     # c1.__add__(c2) 호출됨
result = MyCalculator(result)
result = result + c3

print(result)  # 60

result = c1 * c2   # 200 을 리턴하려면?
print(result)
````

````python
result = c1 + c2 + c3   # 오류
# TypeError: unsupported operand type(s) for +: 'int' and 'MyCalculator'
# c1 + c2 의 결과가 int 타입. c3 은  MyCalculator 타입 객체
````
