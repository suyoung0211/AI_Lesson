### 2. __ str __()
- ì‚¬ìš©ìê°€ ì´í•´í•˜ê¸° ì‰½ê³  ì½ê¸° ì¢‹ì€ í˜•íƒœì˜ ë¬¸ìì—´ í‘œí˜„ì„ ë°˜í™˜í•˜ëŠ” ë©”ì„œë“œë¡œ, ì£¼ë¡œ ê°ì²´ë¥¼ ì¶œë ¥í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.
- __ str __ ëŠ” print() ë‚˜ str() ë¥¼ í†µí•´ í˜¸ì¶œë©ë‹ˆë‹¤.
- repr() ëŠ” ê°ì²´ì˜ ê³µì‹ì ì¸ í‘œí˜„ì„ ì œê³µí•˜ê³ , repr() í•¨ìˆ˜ë¥¼ í†µí•´ í˜¸ì¶œë©ë‹ˆë‹¤.

````python
# __str__ ê³¼ __repr__ ë¹„êµ
class Book:
  def __init__(self,title,writer):
    self.title=title
    self.writer=writer

  # str í•¨ìˆ˜ë¥¼ ì˜¤ë²„ë¼ì´ë”©í•˜ê¸°
  def __str__(self):
   return f"{self.title} : {self.writer}"

  def __repr__(self):
    return f"Book(title='{self.title}', writer='{self.writer}')"

book = Book('ë¯¸ì¹œë“¯ì´ ì¬ë°ŒëŠ” ìë°”','ê¹€ë•¡ë•¡')
print(book)   # ì˜¤ë²„ë¼ì´ë”© í•œ ë¦¬í„´ê°’ì„ ì¶œë ¥í•˜ê¸°. reprê³¼ str ì¤‘ str ì‹¤í–‰
print(str(book))
print(repr(book))
````

### 4. __ getitem __()
- íŒŒì´ì¬ì—ì„œ ì¸ë±ì‹±ì„ ì§€ì›í•˜ê¸° ìœ„í•´ ì‚¬ìš©ë˜ëŠ” íŠ¹ë³„í•œ ë©”ì„œë“œ
- ê°ì²´ì˜ íŠ¹ì • í•­ëª©ì„ ê°€ì ¸ì˜¤ê¸° ìœ„í•´ ëŒ€ê´„í˜¸([])ë¥¼ ì‚¬ìš©í•  ë•Œ í˜¸ì¶œë©ë‹ˆë‹¤.
- ì¸ë±ìŠ¤ë‚˜ í‚¤ì— í•´ë‹¹í•˜ëŠ” ê°’ì„ ë°˜í™˜í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.

````python
# __getitem__() : ê°ì²´ì˜ ì¸ìŠ¤í„´ìŠ¤ë³€ìˆ˜(ì†ì„±)ì„ ì¸ë±ìŠ¤ë¡œ ì ‘ê·¼í•  ìˆ˜ ìˆê²Œ í•œë‹¤.
class Point:
    def __init__(self, x, y,z):
        self.x = x
        self.y = y
        self.z = z

    def __getitem__(self, index):
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        elif index == 2:
            return self.z
        else:
            return -1
pt = Point(5, 3, 7)
print(pt)
print(pt[0])
print(pt[1])
print(pt[2])
print(pt[100])
````

### 5. __ call __()
- íŒŒì´ì¬ì—ì„œ ê°ì²´ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ í˜¸ì¶œí•  ìˆ˜ ìˆê²Œ í•´ì£¼ëŠ” íŠ¹ë³„í•œ ë©”ì„œë“œì…ë‹ˆë‹¤.
- í•´ë‹¹ í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ì¸ìŠ¤í„´ìŠ¤ì— ëŒ€í•´ ê´„í˜¸() ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°’ì„ ì „ë‹¬í•˜ê³  ê²°ê³¼ë¥¼ ë°˜í™˜ ë°›ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
# __call__() ë©”ì†Œë“œ : ê°ì²´ë¥¼ ì°¸ì¡°í•˜ëŠ” ë³€ìˆ˜ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ ì‚¬ìš©í•  ìˆ˜ ìˆê²Œ í•œë‹¤.
#  â†ª ì•„ë˜ ì˜ˆì‹œëŠ” __call__ ë©”ì†Œë“œê°€ action ì¸ìë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.
````

````python
class Dog:
  def __init__(self, name, family):
    self.name = name
    self.family = family
  def __call__(self, action):
    return f"{self.name}, the {self.family}, is now {action}!"

# Dog ê°ì²´ ìƒì„±
my_dog = Dog("Buddy", "Golden Retriever")
print(my_dog)

# __call__ ë©”ì†Œë“œ í˜¸ì¶œ . ì¸ìê°’ action ì „ë‹¬í•¨.
print(my_dog("running"))  # Buddy, the Golden Retriever, is now running!
print(my_dog("barking"))  # Buddy, the Golden Retriever, is now barking!
````

### 6. __ add __() ë©”ì„œë“œ (ì—°ì‚°ì ì˜¤ë²„ë¡œë”©)
- ì´ ì½”ë“œì—ì„œ add()ëŠ” ì—°ì‚°ì ì˜¤ë²„ë¡œë”©(Operator Overloading) ì„ í†µí•´ + ì—°ì‚°ìë¥¼ ì‚¬ìš©ì ì •ì˜í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.

````python
# __add__() ë©”ì†Œë“œ : ê°ì²´ë“¤ì˜ ë§ì…ˆ ì—°ì‚°ì„ ì¬ì •ì˜
# __mul__() ë©”ì†Œë“œ : ê°ì²´ë“¤ì˜ ê³±ì…ˆ ì—°ì‚°ì„ ì¬ì •ì˜
class MyCalculator:
    def __init__(self, num):
        self.num = num  # ìˆ«ì ê°’ì„ ì €ì¥

    def __add__(self, other):
        return self.num + other.num  # ë‘ ê°ì²´ì˜ num ê°’ì„ ë”í•¨

    def __mul__(self, other):
        return self.num * other.num  # ë‘ ê°ì²´ì˜ num ê°’ì„ ê³±í•¨


# ê°ì²´ ìƒì„±
c1 = MyCalculator(10)    # ì´ ê°ì²´ì˜ num ë³€ìˆ˜ê°’ì´ 10
c2 = MyCalculator(20)
c3 = MyCalculator(30)

# + ì—°ì‚° ìˆ˜í–‰
result = c1 + c2     # c1.__add__(c2) í˜¸ì¶œë¨
result = MyCalculator(result)
result = result + c3

print(result)  # 60

result = c1 * c2   # 200 ì„ ë¦¬í„´í•˜ë ¤ë©´?
print(result)
````
