```
# 변수(상수)
PI = 3.141592653589793

# 함수
def print_name():
    print(f'😁__name__ : {__name__}')

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        """원의 넓이 계산: πr²"""
        return PI * (self.radius ** 2)

    def circumference(self):
        """원의 둘레 계산: 2πr"""
        return 2 * PI * self.radius
```
