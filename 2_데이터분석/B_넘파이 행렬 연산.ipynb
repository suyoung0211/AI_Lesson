### 행렬 곱
- 벡터, 스칼라, 행렬 모두 처리 가능
- np.dot() : dot product(내적) 는 1차원 배열(벡터)의 곱셈의 합. 2차원 배열 사용 가능
- np.matmul() : 다차원 배열의 행렬 곱에 특화.
- @ 기호 : matmul(), dot() 함수와 같은 역할.

````python
# 1차원 배열의 내적(product) => 스칼라 값
#    ㄴ dot 메소드 , 기호 @
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
# 내적 연산 : 같은 위치의 값을 각각 곱해서 더하기. 수학적 의미는 벡터 간의 곱셈.
print(np.dot(a, b))   # 1x4 + 2x5 +3x6
````

````python
# 파이썬 3.5 이상 버전. 행렬 곱
print(a @ b)
````

````python
# 2차원 배열 행렬곱은 A행과 B열의 내적으로 구성된다.
ndarr3 = np.array([[1, 2, 3],
                   [2, 3, 4]])
ndarr4 = np.array([[1, 2],
                   [3, 4],
                   [5, 6]])
print('ndarr3 =')
print(ndarr3)
print('ndarr4 =')
print(ndarr4)
print(ndarr3.shape)
print(ndarr4.shape)

# 2차원 배열에서 내적 (다차원은 더 복잡)
print((1*1 + 2*3 + 3*5), (1*2 + 2*4 + 3*6))   # ndarr3 0행 @ ndarr4 0열, ndarr3 0행 @ ndarr4 1열
print((2*1 + 3*3 + 4*5), (2*2 + 3*4 + 4*6))   # ndarr3 1행 @ ndarr4 0열, ndarr3 1행 @ ndarr4 1열
````

````python
# 곱(product, 내적) 연산 정리
r1 = ndarr3 @ ndarr4     #
r2 = np.dot(ndarr3, ndarr4)  #
r3 = np.matmul(ndarr3, ndarr4)  # 다차원 배열에 특화한 행렬 곱(딥러닝, 텐서 연산에 적합)
# matmul : matrix multuply

print (r1)
print(type(r1))
print('--------------------------')
print (r2)
print(type(r2))
print('--------------------------')
print (r3)
print(type(r3))
````

### 요약

```
- np.dot(A,B)                    1D: 내적 / 2D: 행렬곱
- np.matmul(A,B)                 다차원 행렬곱만 수행(다차원 배열의 많은 연산횟수를 효율적으로 수행)
- A @ B                          파이썬 3.5 버전에서 dot, matmul 함수 대신에 사용하는 기호
```

```
A 행렬 : (m,n)
B 행렬 : (x,y)
A와 B의 행렬곱이 가능하려면 n과 x 값이 같아야 하고
A와 B의 행렬곱 결과는 (m,y) shape 입니다.
```

````python
# 2 x 3
ndarr3 = np.array([[1, 2, 3],
                   [2, 3, 4]])
# 2 x 2
ndarr4 = np.array([[1, 2],
                   [3, 4]])
````

````python
# 행렬곱
ndarr3 @ ndarr4
# 오류 : ValueError: matmul: Input operand 1 has a mismatch in its core dimension 0
````

- 3차원 transpose()

````python
# 3차원 transpose
# 3차원 (z,y,x) 는 0,1,2 를
# (x,y,z) 2,1,0
# (y,x,z) 1,2,0
# (x,z,y) 2,0,1
# (z,x,y) 0,2,1
#   옵션은 transpose 위치를 지정 (y,x,z) 이면
#   arr1.transpose(1,2,0)
````

````python
ndarr3 = np.arange(24).reshape(2, 3, 4)  # shape: (2, 3, 4)
ndarr3
````

````python
transposed = ndarr3.transpose(1, 0, 2)
# z와 y축 바꾸기
# shape: (3, 2, 4)
transposed
````

````python
transposed = ndarr3.transpose(0, 2, 1)
# x와 y축 바꾸기
# shape: (2, 3, 4)
transposed
````

- 데이터 자체는 바뀌지 않음: 메모리 상의 값은 그대로이고, 보는 관점만 바뀌는 것.

- 고차원 배열에서 매우 중요: 특히 이미지 처리(CNN), 시계열 분석(RNN), 행렬 연산 등에서 자주 사용됨.

- 축 번호는 0부터 시작: 예를 들어 (batch, channel, height, width) 구조에서 transpose(0, 2, 3, 1)은 채널을 마지막으로 이동시킴.
  + height 는 이미지의 세로
  + width 는 이미지의 가로
  + channel 는 이미지의 r,g,b
  + batch 는 이미지의 갯수
