````python
# csv 파일의 경로를 포함하여 파일 읽어오기
import pandas as pd
filepath = '/content/drive/MyDrive/AI Lesson/KOBIS_월별박스오피스_2025-09-18.csv'
df = pd.read_csv(filepath,encoding='cp949')
df
````

````python
# ↪특정 위치에 값이 없으면 NaN (Not a Number)
#  : 결측치(값). 이 값들은 유지 또는 무시할 값은 삭제
````

- 💥 결측치 행 삭제
`df.dropna()`

  - 행 기준으로 삭제 (기본값)	: df.dropna(axis=0)
  - 하나라도 NaN이면 삭제	: df.dropna(how='any')
  - 모든 값이 NaN일 때만 삭제	: df.dropna(how='all')
  - 특정 열만 기준으로 NaN 검사 : df.dropna(subset=['A', 'B'])
  - 원본 DataFrame 수정	: df.dropna(inplace=True)
  - NaN이 아닌 값이 3개 이상인 행만 유지 : df.dropna(thresh=3)

````python
# 행의 모든 컬럼이 NaN 일 때만 행 삭제
# 위 값에서 3번 행만 해당
df_2 = df.dropna(how='all')
df_2
````

````python
# 0~3번 행을 제외하고 슬라이싱
# 컬럼 제목부터 가져오기
df_2 = df_2[4:]
df_2
# 아래 결과에서 현재 컬럼은 Unnamed. 컬럼은 교체 합니다.
````

````python
# 컬럼명의 공백 제거하기
new_columns = [x.strip() for x in df_2.iloc[0].values]
new_columns[4]='매출액점유율'
new_columns
````

- 💥 데이터 프레임 행/열 제거
`df.drop()`
  - drop()은 라벨(label)을 기준으로 행 또는 열을 삭제합니다.
  - 삭제할 대상은 인덱스(행) 또는 열 이름으로 지정합니다.
  - 원본을 수정할지 여부는 inplace=True로 결정합니다.

  - 속성
    ```
      labels	: 삭제할 행/열의 이름 또는 리스트	['A', 'B']
      axis	  : 삭제 방향: 0은 행, 1은 열	axis=1 → 열 삭제
      index	  : 삭제할 행의 인덱스 위치. 범위값 또는 값 나열합니다.	index=[0, 2]
      columns	: 삭제할 열 이름	columns=['A']
      inplace	: 원본 수정 여부 (True면 직접 수정)	inplace=True
      errors	: 삭제할 라벨이 없을 때 처리 방식 ('ignore' 또는 'raise')  errors='ignore'
    ```  

````python
# df_2[3:] 슬라이싱 대신에, 0과1 위치의 행을 삭제
df_2.drop(df_2.index[0:2],inplace=True)
df_2
````

````python
# 맨 마지막 행(합계 행) 삭제
df_2.drop(df_2.index[-1],inplace=True)
````

````python
# 인덱스 초기화
df_2.reset_index(drop=True,inplace=True)
df_2
````

### top 100개 추출해서 통계 분석하기
- 위의 데이터에서 하위권 데이터는 의미가 없다고 판단이 됩니다.
- 주요한 분석 대상인 컬럼의 값이 매우 낮다. 거의 0에 가깝다.

````python
# 위에서 부터 100개 추출
df_top100 = df_2.head(100)
df_top100
````

- 수치 데이터가 모두 문자열 입니다. 관계식으로 인덱싱 못합니다.
  + 데이터의 형식을 변환합니다.

- 매출액,매출액점유율,누적매출액,관객수,누적관객수,스크린수,상영횟수 : 수치형 컬럼
  + 평균, 표준편차 등 수치형 데이터의 통계를 구하기 위해 변환이 필요함.

````python
df_top100.columns
````

````python
# 정수로 변환할 컬럼
num_cols = ['순위','매출액', '누적매출액', '관객수', '누적관객수', '스크린수', '상영횟수']
# 실수로 변환 컬럼은 '매출액점유율'
````

````python
df_top100['매출액']   # dtype: object -> 문자열입니다.
````

- 문자열 -> 정수

````python
# 방법1️⃣ to_numeric() : 문자열을 float64로 변환. 변환하기전에 숫자외의 문자는 제거해야 함.
#                        문자열 -> 실수 -> 정수
# errors='coerce' 는 to_numeric 중 오류 생기면 NaN 저장
# downcast='integer' 는 실수를 정수로 다운캐스팅

for col in num_cols:
  df_top100[col] = pd.to_numeric(df_top100[col].str.replace(r'[,%]','',regex=True),errors='coerce',downcast='integer')

# 방법2️⃣ df_top100[num_cols].apply(람다식)
````

- 문자열 -> 실수

````python
# 방법1️⃣
# df_top100['매출액점유율'] = pd.to_numeric(df_top100['매출액점유율'].str.replace('%',''),errors='coerce')

# 방법2️⃣  : astype() 의 인자는 'float', 'int', 'str'
df_top100['매출액점유율'] = df_top100['매출액점유율'].str.replace('%','').astype(float)
````
